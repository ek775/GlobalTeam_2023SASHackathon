# Exploring Schemas on data.healthcare.gov

```python
#imports
import requests
import json
import pandas as pd
```

```python
#test connection
gov_data_API = requests.get("https://data.healthcare.gov/api/1/metastore/schemas")
print(gov_data_API.status_code)
```

```python
#obtain schemas
schemas = gov_data_API.text
schemas = json.loads(schemas)
schemas
```

```python
#schemas for all datasets
datasets = requests.get("https://data.healthcare.gov/api/1/metastore/schemas/dataset/items")
print(datasets.status_code)
dst = datasets.text
dst = json.loads(dst)
dst
```

```python
datasets_info = pd.DataFrame(dst)
datasets_info
```
